```
def price_category(price):
    if price < 15000:
        return "Budget"
    elif price <= 40000:
        return "Mid-range"
    else:
        return "Premium"

df["price_category"] = df["price"].apply(price_category)

print(df[["model","price","price_category"]].head())
```

```
def check_5g(x):
    if x == 1:
        return "Future Ready"
    else:
        return "Standard Device"

df["5g_status"] = df["has_5g"].apply(check_5g)

print(df[["model","has_5g","5g_status"]].head())
```

```
def battery_type(battery):
    if battery >= 5000:
        return "High Battery"
    elif battery >= 4000:
        return "Medium Battery"
    else:
        return "Low Battery"

df["battery_category"] = df["battery_capacity"].apply(battery_type)

print(df[["model","battery_capacity","battery_category"]].head())
```

```
def performance_category(row):
    if row["ram_capacity"] >= 8 and row["internal_memory"] >= 128:
        return "High Performance"
    else:
        return "Normal Performance"

df["performance"] = df.apply(performance_category, axis=1)

print(df[["model","ram_capacity","internal_memory","performance"]])
```

```
def price_count(price):

    count = 0

    for i in df['price']:
        if i > price:
            count += 1

    print(count)

price_count(50000)
```

```
def battery_phone(value):

    for i in range(len(df)):
        if df['battery_capacity'][i] > value:
            print(df['model'][i])

battery_phone(5000)
```

```
def avg_price():

    total = 0

    for price in df['price']:
        total += price

    return total / len(df)

print(avg_price())
```

```
def ram_phone():

    phones = []

    for i in range(len(df)):
        if df['ram_capacity'][i] >= 8:
            phones.append(df['model'][i])

    return phones

print(ram_phone())
```

```
def expensive_phone():

    max_price = 0
    phone = ""

    for i in range(len(df)):

        if df['price'][i] > max_price:
            max_price = df['price'][i]
            phone = df['model'][i]

    return phone, max_price

print(expensive_phone())
```

```
def phone_5g():

    g5 = 0
    non5g = 0

    for i in df['has_5g']:

        if i == True:
            g5 += 1
        else:
            non5g += 1

    return g5, non5g

print(phone_5g())
```

```
class Smartphone:

    def __init__(self, brand, price, ram, battery):
        self.brand = brand
        self.price = price
        self.ram = ram
        self.battery = battery

    def is_premium(self):
        return self.price > 40000

    def battery_category(self):

        if self.battery > 5000:
            return "High"

        elif self.battery >= 4000:
            return "Medium"

        else:
            return "Low"


p1 = Smartphone("Samsung",45000,8,5000)
p2 = Smartphone("Apple",90000,6,3200)

print(p1.is_premium())
print(p2.battery_category())
```
